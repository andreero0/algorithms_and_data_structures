#### Starter Code for Question 2

```python
# Import List from typing for type hinting
from typing import List
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    # If the root is None, return an empty list
    if not root:
        return []
    
    # Initialize the result list to store all paths
    paths = []
    
    # Helper function to perform depth-first search
    def dfs(node, current_path):
        # If the node is None, return
        if not node:
            return
        
        # Add current node's value to the path
        current_path.append(node.val)
        
        # If this is a leaf node (no children), add the path to results
        if not node.left and not node.right:
            paths.append(current_path.copy())
        
        # Recursively explore left and right subtrees
        if node.left:
            dfs(node.left, current_path)
        if node.right:
            dfs(node.right, current_path)
        
        # Backtrack by removing the last node
        current_path.pop()
    
    # Start the depth-first search from the root
    dfs(root, [])
    
    return paths
```


-   Code the solution to your assigned problem in Python (code chunk). Try to find the best time and space complexity solution!


```python
from typing import List

# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    # If the root is None, return an empty list
    if not root:
        return []
    
    # Initialize the result list to store all paths
    paths = []
    
    # Helper function to perform depth-first search (DFS)
    def dfs(node, current_path):
        if not node:
            return
        
        # Add current node's value to the current path
        current_path.append(node.val)
        
        # If this is a leaf node (no left or right child) add a copy of the path to results
        if not node.left and not node.right:
            paths.append(current_path.copy())
        else:
            # Recursively explore the left and right subtrees
            dfs(node.left, current_path)
            dfs(node.right, current_path)
        
        # Backtrack by removing the last node added
        current_path.pop()
    
    # Start DFS from the root
    dfs(root, [])
    return paths

# Optional test function to verify the solution
def test_bt_path():
    # Test Case 1: Balanced tree
    #      1
    #    /   \
    #   2     3
    #  / \
    # 4   5
    root1 = TreeNode(1)
    root1.left = TreeNode(2)
    root1.right = TreeNode(3)
    root1.left.left = TreeNode(4)
    root1.left.right = TreeNode(5)
    print("Test Case 1 Paths:", bt_path(root1))
    
    # Test Case 2: Unbalanced tree
    #      5
    #     / \
    #    3   7
    #       /
    #      6
    root2 = TreeNode(5)
    root2.left = TreeNode(3)
    root2.right = TreeNode(7)
    root2.right.left = TreeNode(6)
    print("Test Case 2 Paths:", bt_path(root2))
    
    # Test Case 3: Single Node
    root3 = TreeNode(42)
    print("Test Case 3 Paths:", bt_path(root3))

# Run the tests (if required)
test_bt_path()
```


-   Explain why your solution works


Your answer here: 

The solution works through a depth-first search (DFS) strategy:

1. Start at the root node
2. Build a path as we traverse down the tree
3. When we reach a leaf (node with no children):
   - Create a copy of the current path
   - Add this path to our results
4. Use recursion to explore all possible paths
5. Backtracking (popping the last node) allows us to explore 
   alternative routes without losing previous paths

Key Mechanisms:
- Recursion explores all branches
- .copy() prevents path overwriting
- current_path.pop() allows backtracking


-   Explain the problem’s time and space complexity


Your answer here

Time Complexity: O(N)
- Each node in the tree is visited exactly once during the DFS traversal
- At each node, we perform constant time operations:
    - Appending to current_path: O(1)
    - Creating a copy of the path for leaf nodes: O(1)
    - Popping from current_path during backtracking: O(1)
- Total operations = number of nodes × constant time operations
- Therefore, overall time complexity is O(N), where N is the number of nodes

Space Complexity: O(H)
- The space complexity is determined by two factors:
    - Recursion Stack: Maximum depth of recursion equals tree height (H)
    - Path Storage: Each path can be at most H nodes long
- For a balanced tree: H = log N (best case)
- For a skewed tree: H = N (worst case)
- Therefore:
    - Best case: O(log N)
    - Worst case: O(N)

Factors Affecting Complexity:
- Number of nodes
- Tree structure (balanced vs unbalanced)
- Depth of the tree
