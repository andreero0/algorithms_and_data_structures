````python
# Your answer here
# Definition for a binary tree node.
from collections import deque

class TreeNode(object):
     def __init__(self, val = 0, left = None, right = None):
         self.val = val
         self.left = left
         self.right = right
def is_duplicate(root: TreeNode) -> int:
    visited = set()
    queue = deque([root])

    while queue:
        current_node = queue.popleft()

        if current_node is None:
            continue
        
        if current_node.val in visited:
            return current_node.val
        
        else:
            visited.add(current_node.val)
            queue.append(current_node.left)
            queue.append(current_node.right)
    
    return -1
````


-   Explain why their solution works in your own words.


Solution Works Because:

5. Handles edge cases like empty trees or no duplicates

Key Mechanisms:
- Queue ensures systematic tree exploration
- Set provides O(1) lookup for seen values
- Null node handling prevents errors
- Early return optimizes performance


-   Explain the problem’s time and space complexity in your own words.


Time Complexity: O(N)

- Queue and set operations are O(1)
- Total time proportional to number of nodes

Space Complexity: O(W)
- W = Maximum width of the tree
- Queue stores nodes at current level
- Set stores unique values seen so far
- Worst case: storing all nodes at widest level

```python
def improved_is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1
    visited = {}  # value:level dictionary
    queue = deque([(root, 0)])  # (node, level)
    min_level = float('inf')

    while queue:
        current_node, level = queue.popleft()
```
